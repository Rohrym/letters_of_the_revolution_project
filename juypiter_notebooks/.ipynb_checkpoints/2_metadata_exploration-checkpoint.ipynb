```python
#loading the relevant libraries
import pandas as pd
import datetime
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr
from heapq import nlargest
import math
```

```python
df = pd.read_csv('/content/drive/Shareddrives/Letters of the Revolution Project/Repo: Revolutionary War/Data/Archive data/founders_online_archive_revwar_cooked.csv', index_col=0)
```

```python
#changing datatypes and adding a tokenized text column
df[['date_from', 'date_to']] = df[['date_from', 'date_to']].astype('datetime64[ns]')
split_authors_recipients = df[['authors','recipients','id','date_from']].set_index(['id','date_from']).apply(lambda x: x.str.split('|')).explode('authors').explode('recipients').reset_index() 
split_authors_recipients['authors'] = split_authors_recipients['authors'].str.strip()
split_authors_recipients['recipients'] = split_authors_recipients['recipients'].str.strip()                                                                                                  
```

# Data exploration

## Letter counts

```python
#saving a variable with list of founding fathers
founders = ['Washington, George', 'Adams, John', 'Jefferson, Thomas', 
              'Madison, James', 'Hamilton, Alexander', 'Franklin, Benjamin','Jay, John']
```

```python
#total amount of correspondence
print('amount of correspondence:', len(df.index))
```

```python
print('amount of authors:', len(split_authors_recipients['authors'].unique()))
print('amount of recipients:', len(split_authors_recipients['recipients'].unique()))
```

```python
#average amount of letters authored per person
print('avg authored per person:',(len(df.index))/(len(split_authors_recipients['authors'].unique())))
print('avg received per person:',(len(df.index))/(len(split_authors_recipients['recipients'].unique())))
```

```python
#amount of the corpus authored by one of the founding fathers
p = len(split_authors_recipients.loc[split_authors_recipients['authors'].isin(founders)])/(len(df.index))
print('authored:', f"{p:.2%}")

#amound of the corpus received by one of the founding fathers
p = len(split_authors_recipients.loc[split_authors_recipients['recipients'].isin(founders)])/(len(df.index))
print('received:',f"{p:.2%}")
```

```python
#amount of letters exchanged between the founding fathers
l_exchanged = len(split_authors_recipients.loc[split_authors_recipients['authors'].isin(founders) & split_authors_recipients['recipients'].isin(founders)])
print('amount exchanged:',l_exchanged)

#amount as a percentage of the archive
p = l_exchanged/(len(df.index))
print('percentage total:', f"{p:.2%}")
```

```python
#the top 10 authors and the amount of correspondence (co-)written
amount_authored = split_authors_recipients['authors'].value_counts()
print(amount_authored.head(10))
```

```python
#amount of authors that (co-)authored 1 to 10 letters.
values_amount_authored = amount_authored.value_counts()
print(values_amount_authored.sort_index().head(10))
```

```python
p = values_amount_authored.sort_index().head(10).sum()/(len(split_authors_recipients['authors'].unique()))
print('percentage that authored 1 to 10 documents:',f"{p:.2%}")
```

```python
#the top 10 recipients and the amount of correspondence received
amount_received = split_authors_recipients['recipients'].value_counts()
print(amount_received.nlargest(10))
```

```python
#amount amount of recipientes that received 1 to 10 letters.
values_amount_received = amount_received.value_counts()
print(values_amount_received.sort_index().head(10))
```

```python
p = (values_amount_received.sort_index().head(10).sum())/(len(df['recipients'].unique()))
print('percentage that received 1 to 10 documents:',f"{p:.2%}")
```

```python
amount_auth_and_rec = pd.merge(amount_authored,amount_received, left_index=True,right_index=True).fillna(0)
amount_auth_and_rec['combined'] = (amount_auth_and_rec['authors'] + amount_auth_and_rec['recipients'])
amount_auth_and_rec.sort_values('combined', ascending=False).head(10)
```

```python
ax = sns.scatterplot(data = amount_auth_and_rec, x='authors', y='recipients')
ax.set_yscale('log')
ax.set_xscale('log')
plt.title("Amount of authors and recipients (logscale)")
plt.show()

print('Correlation coefficient and p-value: \n')
print(pearsonr(amount_auth_and_rec['authors'],amount_auth_and_rec['recipients']))
```

```python
sns.boxplot(data = amount_auth_and_rec[['authors','recipients']], showfliers = False)
plt.title("Boxplot of occurence of lettercounts without outliers")
plt.show()
```

## Counts over time

```python
ax = df['date_from'].dt.to_period('M').value_counts().sort_index(ascending=True).plot(title='Amount of correspondence per month')
plt.show()
```

```python
#correspondence per founder, per year.
for founder in founders:
    r = random.random()
    b = random.random()
    g = random.random()
    color = (r, g, b)
    
    dft = split_authors_recipients.loc[split_authors_recipients['authors'] == founder]
    ax = dft['date_from'].dt.to_period('M').value_counts().sort_index(ascending=True).plot(title=founder, color=color) 
    plt.show()
```
